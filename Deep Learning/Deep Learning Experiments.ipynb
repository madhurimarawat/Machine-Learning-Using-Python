## üî¨ **Experiment 1: Introduction to Neural Networks and Training a Simple Perceptron**  
### This experiment introduces the fundamental concepts of neural networks by training a simple perceptron model. The perceptron is applied to classify basic linearly separable data, providing insights into how weights and biases adjust through iterative learning.

### üß† What is a Perceptron?  

A **perceptron** is a computational unit that mimics a biological neuron üß¨. It takes several inputs, multiplies them with weights, sums them, adds a bias, and passes the result through an **activation function** (usually a step or sigmoid function).

üßÆ **Mathematically**:  
- `z = w1*x1 + w2*x2 + ... + wn*xn + b`  
  ‚Üí which is: `z = (w ‚Ä¢ x) + b`  
- Output `y` is:  
  - `1`, if `z >= 0`  
  - `0`, otherwise

Where:  
- `x` = input vector  
- `w` = weight vector  
- `b` = bias  
- `y` = output prediction

üéØ The goal of training a perceptron is to find the best weights (`w`) and bias (`b`) to separate two classes correctly.

---

### üîÑ How the Perceptron Learns  

The perceptron updates its weights using the **Perceptron Learning Rule** when it makes a mistake:

üìê **Update Rule**:  
- `w_new = w_old + Œ∑ * (y_true - y_pred) * x`  
- `b_new = b_old + Œ∑ * (y_true - y_pred)`

Here,  
- `Œ∑` (eta) is the **learning rate** (a small number like 0.01)  
- `y_true` is the actual label  
- `y_pred` is the predicted label  

üß© This update nudges the model in the right direction and continues until it classifies all data correctly or stops after many tries.

---

### üèóÔ∏è Moving Beyond: Multi-Layer Perceptron (MLP)

A single perceptron only works when data is **linearly separable**. But real-world problems are rarely that simple. So we use **Multi-Layer Perceptrons (MLPs)** ü§ñ.

An **MLP** is a network made up of:  
- An **input layer**  
- One or more **hidden layers**  
- An **output layer**

Each neuron connects to all neurons in the next layer (this is called a **dense** or **fully connected** layer).

üì∏ **MLP Architecture Example**  
![Multi-Layer Perceptron Diagram](https://www.researchgate.net/publication/334609713/figure/fig1/AS:783455927406593@1563801857102/Multi-Layer-Perceptron-MLP-diagram-with-four-hidden-layers-and-a-collection-of-single.jpg)

---

### üîÅ What is Backpropagation?

**Backpropagation** is the learning algorithm used in MLPs to adjust weights from **end to start** (backward). It helps the network **learn from its errors**.

üìâ **Simple Steps**:  
1. Make a prediction with current weights  
2. Calculate the error using the **loss function**:  
   - `Loss = 0.5 * (y_true - y_pred)^2`  
3. Use **gradient descent** to adjust weights by figuring out how much each weight contributed to the error

üìê **Gradient Descent Rule**:  
- `w = w - Œ∑ * (dLoss/dw)`

This is repeated many times (called **epochs**) until the loss becomes small and predictions become accurate üìä.

---

### üß† Summary Table

| Concept             | Role in Neural Network                        |
|---------------------|------------------------------------------------|
| Perceptron          | Basic unit of computation                    |
| Activation Function | Decides whether neuron fires (like Step)      |
| Learning Rule       | Updates weights when prediction is wrong      |
| MLP                 | Uses multiple layers to handle complex data   |
| Backpropagation     | Adjusts weights layer-by-layer using errors   |

---

### üìå Final Note

This experiment builds the **foundation** of deep learning üí°. The perceptron teaches us how weights and biases work. MLP and backpropagation show how we can **scale this learning** to bigger, smarter networks that tackle real-world problems like image recognition üñºÔ∏è, language understanding üó£Ô∏è, and even self-driving cars üöó.

```python
# Experiment 1: Perceptron for Linearly Separable Data
# Author: Madhurima Rawat

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set pastel theme for graphs
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üîç Experiment 1: Training a Simple Perceptron")
print("------------------------------------------------")
print("This experiment demonstrates how a perceptron can classify linearly separable data.")
print("The perceptron learns by updating its weights based on prediction errors.")
print("It is a binary classifier using a step activation function.\n")

# Generate a simple linearly separable dataset
def generate_data():
    np.random.seed(0)
    # Class 0: Points centered around (2, 2)
    class_0 = np.random.randn(50, 2) + [2, 2]
    # Class 1: Points centered around (6, 6)
    class_1 = np.random.randn(50, 2) + [6, 6]

    # Combine data and create labels
    X = np.vstack((class_0, class_1))
    y = np.array([0]*50 + [1]*50)

    return X, y

# Define Perceptron class
class Perceptron:
    """
    A simple implementation of the Perceptron algorithm.
    It learns a linear decision boundary for binary classification.
    """
    def __init__(self, input_size, learning_rate=0.01, epochs=100):
        self.weights = np.zeros(input_size + 1)  # +1 for bias
        self.lr = learning_rate
        self.epochs = epochs
        self.errors = []

    def activation_fn(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation_fn(z)

    def fit(self, X, y):
        print("üìö Starting training...\n")
        for epoch in range(self.epochs):
            error_count = 0
            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                update = self.lr * (target - prediction)
                self.weights[1:] += update * xi
                self.weights[0] += update  # bias
                error_count += int(update != 0.0)
            self.errors.append(error_count)
            print(f"Epoch {epoch+1:2d} ‚Üí Misclassifications: {error_count}")
        print("\n‚úÖ Training complete.\n")
        print(f"üìå Final Weights: {self.weights[1:]}")
        print(f"üìå Final Bias: {self.weights[0]}")
        print(f"üìà Total Epochs: {self.epochs}")
        print(f"üìä Final Training Error: {self.errors[-1]}")
        # For Newline
        print()

# Plotting function to visualize decision boundary
def plot_decision_boundary(X, y, model):
    """
    Visualizes the decision boundary learned by the perceptron.
    """
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = np.array([model.predict(np.array([x, y])) for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="Pastel1")
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="pastel", edgecolor="black")
    plt.title("üîç Perceptron Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(title="Class")
    plt.show()

# Plot training error over epochs
def plot_errors(errors):
    """
    Plots the number of misclassifications per epoch.
    """
    plt.figure(figsize=(8, 4))
    sns.lineplot(x=range(len(errors)), y=errors, color="skyblue", marker="o")
    plt.title("üìâ Training Error Over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Number of Misclassifications")
    plt.show()

# Main execution
if __name__ == "__main__":
    # Step 1: Generate data
    X, y = generate_data()

    # Step 2: Initialize and train perceptron
    perceptron = Perceptron(input_size=2, learning_rate=0.1, epochs=20)
    perceptron.fit(X, y)

    # Step 3: Plot decision boundary
    plot_decision_boundary(X, y, perceptron)

    # For Newline
    print()

    # Step 4: Plot training errors
    plot_errors(perceptron.errors)
```

## üîÅ **Experiment 2: Implementing Feedforward Neural Networks (FNN)**  
### A Feedforward Neural Network (FNN) is implemented and trained using the backpropagation algorithm. The model performs classification tasks and helps in understanding the core workflow of a multilayer perceptron.

```python
# Experiment 2: Implementing Feedforward Neural Networks (FNN)
# Author: Madhurima Rawat

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set pastel theme for graphs
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üîÅ Experiment 2: Implementing Feedforward Neural Network")
print("--------------------------------------------------------")
print("This experiment demonstrates a basic Feedforward Neural Network (FNN) using backpropagation.")
print("FNNs are composed of layers of neurons where each neuron uses an activation function to produce output.")
print("Backpropagation is used to update weights and minimize the classification error.\n")

# Generate dataset (same as before for comparison)
def generate_data():
    np.random.seed(1)
    class_0 = np.random.randn(50, 2) + [2, 2]
    class_1 = np.random.randn(50, 2) + [6, 6]
    X = np.vstack((class_0, class_1))
    y = np.array([0]*50 + [1]*50)
    return X, y

# Activation and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Define Feedforward Neural Network class
class FNN:
    """
    A simple implementation of a Feedforward Neural Network with one hidden layer.
    """
    def __init__(self, input_size, hidden_size, output_size, lr=0.1, epochs=100):
        self.lr = lr
        self.epochs = epochs
        self.errors = []
        # Initialize weights
        self.w1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.w2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def fit(self, X, y):
        print("üß† Starting training...\n")
        y = y.reshape(-1, 1)
        for epoch in range(self.epochs):
            # Forward pass
            z1 = np.dot(X, self.w1) + self.b1
            a1 = sigmoid(z1)
            z2 = np.dot(a1, self.w2) + self.b2
            a2 = sigmoid(z2)

            # Compute error
            loss = np.mean((y - a2) ** 2)
            self.errors.append(loss)

            # Backpropagation
            d_a2 = (a2 - y)
            d_z2 = d_a2 * sigmoid_derivative(a2)
            d_w2 = np.dot(a1.T, d_z2)
            d_b2 = np.sum(d_z2, axis=0, keepdims=True)

            d_a1 = np.dot(d_z2, self.w2.T)
            d_z1 = d_a1 * sigmoid_derivative(a1)
            d_w1 = np.dot(X.T, d_z1)
            d_b1 = np.sum(d_z1, axis=0)

            # Update weights
            self.w1 -= self.lr * d_w1
            self.b1 -= self.lr * d_b1
            self.w2 -= self.lr * d_w2
            self.b2 -= self.lr * d_b2

            print(f"Epoch {epoch+1:2d} ‚Üí Loss: {loss:.4f}")
        print("\n‚úÖ Training complete.\n")
        print(f"üìà Final Loss: {self.errors[-1]:.4f}")
        print(f"üìâ Total Epochs: {self.epochs}\n")

    def predict(self, X):
        a1 = sigmoid(np.dot(X, self.w1) + self.b1)
        a2 = sigmoid(np.dot(a1, self.w2) + self.b2)
        return (a2 > 0.5).astype(int).flatten()

# Plot decision boundary
def plot_decision_boundary(X, y, model):
    """
    Visualizes the decision boundary learned by the FNN.
    """
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = np.array([model.predict(np.array([[x, y]])) for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="Pastel2")
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="pastel", edgecolor="black")
    plt.title("üîÅ Feedforward Neural Network Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(title="Class")
    plt.show()

# Plot training loss
def plot_loss(losses):
    """
    Plots the loss over epochs.
    """
    plt.figure(figsize=(8, 4))
    sns.lineplot(x=range(len(losses)), y=losses, color="orchid", marker="o")
    plt.title("üìâ Training Loss Over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Mean Squared Error (Loss)")
    plt.show()

# Main execution
if __name__ == "__main__":
    # Step 1: Generate data
    X, y = generate_data()

    # Step 2: Initialize and train the FNN
    fnn = FNN(input_size=2, hidden_size=5, output_size=1, lr=0.1, epochs=50)
    fnn.fit(X, y)

    # Step 3: Plot decision boundary
    plot_decision_boundary(X, y, fnn)

    # For Newline
    print()

    # Step 4: Plot loss
    plot_loss(fnn.errors)
```

## üñºÔ∏è **Experiment 3: Convolutional Neural Networks (CNN) for Image Classification**  
### This experiment focuses on implementing a Convolutional Neural Network (CNN) to classify images from the CIFAR-10 dataset. It demonstrates how convolutional layers extract spatial features and improve performance on image data.

```python
# Experiment 3: Convolutional Neural Networks (CNN) for Image Classification
# Author: Madhurima Rawat

# Import required libraries
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üñºÔ∏è Experiment 3: CNN for Image Classification")
print("------------------------------------------------------")
print("This experiment demonstrates the use of a Convolutional Neural Network (CNN)")
print("to classify images from the CIFAR-10 dataset. CNNs are ideal for extracting")
print("spatial features and improving image classification accuracy.\n")

# Load and preprocess data
def load_data():
    (train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
    train_images, test_images = train_images / 255.0, test_images / 255.0  # Normalize
    class_names = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
                   'Dog', 'Frog', 'Horse', 'Ship', 'Truck']
    return train_images, train_labels, test_images, test_labels, class_names

# Build CNN model
def build_model():
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10))  # 10 classes
    return model

# Plot sample images from dataset
def plot_sample_images(images, labels, class_names):
    plt.figure(figsize=(10, 5))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i])
        plt.xlabel(class_names[labels[i][0]])
    plt.suptitle("üéØ Sample CIFAR-10 Images")
    plt.tight_layout()
    plt.show()
    print()

# Plot training history
def plot_training_history(history):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    sns.lineplot(x=range(len(history.history['accuracy'])), y=history.history['accuracy'], label='Train Acc', marker="o")
    sns.lineplot(x=range(len(history.history['val_accuracy'])), y=history.history['val_accuracy'], label='Val Acc', marker="o")
    plt.title("üìà Model Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")

    plt.subplot(1, 2, 2)
    sns.lineplot(x=range(len(history.history['loss'])), y=history.history['loss'], label='Train Loss', marker="o")
    sns.lineplot(x=range(len(history.history['val_loss'])), y=history.history['val_loss'], label='Val Loss', marker="o")
    plt.title("üìâ Model Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")

    plt.tight_layout()
    plt.show()
    print()

# Main execution
if __name__ == "__main__":
    # Step 1: Load and display data
    train_images, train_labels, test_images, test_labels, class_names = load_data()
    plot_sample_images(train_images, train_labels, class_names)

    # Step 2: Build and compile model
    model = build_model()
    model.compile(optimizer='adam',
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])

    # Step 3: Train the model
    print("üöÄ Starting training...\n")
    history = model.fit(train_images, train_labels, epochs=4,
                        validation_data=(test_images, test_labels), verbose=2)
    print("\n‚úÖ Training complete.\n")

    # Step 4: Plot training history
    plot_training_history(history)
```

## üìà **Experiment 4: Implementing a Recurrent Neural Network (RNN) for Sequence Prediction**  
### An RNN model is developed to handle sequential data, particularly for time series forecasting. This experiment highlights the importance of retaining contextual information over sequences using looped connections.

```python
# Experiment 4: Implementing a Recurrent Neural Network (RNN) for Sequence Prediction
# Author: Madhurima Rawat

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üìà Experiment 4: RNN for Sequence Prediction")
print("--------------------------------------------------------")
print("This experiment builds a simple RNN to predict future values in a time series.")
print("RNNs are designed to retain information from previous time steps, making them")
print("ideal for handling sequential data like stock prices, weather patterns, etc.\n")

# Generate synthetic sequential data (e.g., sine wave)
def generate_sequence():
    x = np.linspace(0, 50, 500)
    y = np.sin(x)
    return x, y

# Prepare data for RNN input (sliding window)
def prepare_data(y, window_size=10):
    X, Y = [], []
    for i in range(len(y) - window_size):
        X.append(y[i:i+window_size])
        Y.append(y[i+window_size])
    X = np.array(X)
    Y = np.array(Y)
    return X.reshape(-1, window_size, 1), Y

# Plot original sequence
def plot_original_sequence(x, y):
    plt.figure(figsize=(10, 4))
    plt.plot(x, y, color='skyblue')
    plt.title("üîÅ Original Sine Wave Sequence")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.show()
    print()

# Plot predictions vs actual
def plot_predictions(y_true, y_pred, title="üîÆ Predicted vs Actual Sequence"):
    plt.figure(figsize=(10, 4))
    plt.plot(y_true, label='Actual', color='lightgreen')
    plt.plot(y_pred, label='Predicted', color='mediumblue')
    plt.title(title)
    plt.xlabel("Time Steps")
    plt.ylabel("Value")
    plt.legend()
    plt.tight_layout()
    plt.show()
    print()

# Main execution
if __name__ == "__main__":
    # Step 1: Generate and visualize sequence
    x, y = generate_sequence()
    plot_original_sequence(x, y)

    # Step 2: Prepare data
    window_size = 20
    X, Y = prepare_data(y, window_size)

    # Step 3: Build RNN model
    model = Sequential([
        SimpleRNN(10, activation='tanh', input_shape=(window_size, 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')

    # Step 4: Train model
    print("üöÄ Starting training...\n")
    model.fit(X, Y, epochs=20, verbose=1)
    print("\n‚úÖ Training complete.\n")

    # Step 5: Make predictions
    y_pred = model.predict(X)

    # Step 6: Plot predictions
    plot_predictions(Y, y_pred.flatten())
```

## üí¨ **Experiment 5: Long Short-Term Memory (LSTM) Network for Sentiment Analysis**  
### A Long Short-Term Memory (LSTM) network is implemented for sentiment analysis using a labeled text dataset. This model showcases how LSTMs manage long-term dependencies in sequences and outperform basic RNNs in many NLP tasks.

```python
# Experiment 5: Long Short-Term Memory (LSTM) Network for Sentiment Analysis
# Author: Madhurima Rawat

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üí¨ Experiment 5: LSTM for Sentiment Analysis")
print("----------------------------------------------------------")
print("This experiment implements an LSTM network to classify movie reviews from the IMDb dataset.")
print("LSTMs are a special kind of RNN capable of learning long-term dependencies, ideal for text data.")
print("They help improve accuracy by reducing the vanishing gradient problem found in standard RNNs.\n")

# Step 1: Load IMDb dataset (pre-tokenized into word indices)
print("üì• Loading IMDb dataset...")
max_features = 100  # Vocabulary size
max_len = 20         # Sequence length (truncate or pad to this length)
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Step 2: Pad sequences for uniform input shape
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
print("‚úÖ Data loaded and padded.\n")

# Step 3: Build LSTM model
model = Sequential([
    Embedding(input_dim=max_features, output_dim=64, input_length=max_len),
    LSTM(64),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 4: Train the model
print("üöÄ Starting training...\n")
history = model.fit(X_train, y_train, epochs=3, batch_size=128, validation_split=0.2, verbose=1)
print("\n‚úÖ Training complete.\n")

# Plot accuracy and loss
def plot_training_metrics(history):
    plt.figure(figsize=(10, 4))
    plt.plot(history.history['accuracy'], label='Train Accuracy', color='skyblue')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy', color='mediumblue')
    plt.title("üìà Training vs Validation Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()
    print()

    plt.figure(figsize=(10, 4))
    plt.plot(history.history['loss'], label='Train Loss', color='lightcoral')
    plt.plot(history.history['val_loss'], label='Val Loss', color='crimson')
    plt.title("üìâ Training vs Validation Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
    print()

# Step 5: Evaluate and plot
plot_training_metrics(history)

# Step 6: Evaluate on test data
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"üìä Test Accuracy: {accuracy:.4f}")
print(f"üìâ Test Loss: {loss:.4f}")
```

## üé® **Experiment 6: Generative Adversarial Networks (GANs) for Image Generation**  
### This experiment involves building a Generative Adversarial Network (GAN) for generating images similar to the ones in the training set, such as MNIST. The adversarial training setup (generator vs. discriminator) is explored in detail.


```python
# Experiment 6: Generative Adversarial Networks (GANs) for Image Generation
# Author: Madhurima Rawat

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.layers import Dense, LeakyReLU, Reshape, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
import warnings
warnings.filterwarnings('ignore')

# Visualization style
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üé® Experiment 6: GANs for Image Generation")
print("------------------------------------------------------------")
print("This experiment builds a simple GAN using the MNIST dataset.")
print("The generator creates fake images, and the discriminator tries to detect them.")
print("Through adversarial training, the generator learns to produce realistic digits.\n")

# Step 1: Load and normalize MNIST data
(X_train, _), _ = mnist.load_data()
X_train = (X_train.astype(np.float32) - 127.5) / 127.5

# Step 2: Build generator
def build_generator():
    model = Sequential([
        Dense(128, input_dim=100),
        LeakyReLU(0.2),
        Dense(784, activation='tanh'),
        Reshape((28, 28))
    ])
    return model

# Step 3: Build discriminator
def build_discriminator():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128),
        LeakyReLU(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Step 4: Build and compile GAN
generator = build_generator()
discriminator = build_discriminator()
discriminator.trainable = False

gan = Sequential([generator, discriminator])
gan.compile(optimizer='adam', loss='binary_crossentropy')

# Step 5: Training loop
def train_gan(epochs=1000, batch_size=128):
    real_labels = np.ones((batch_size, 1))
    fake_labels = np.zeros((batch_size, 1))

    for epoch in range(1, epochs + 1):
        # Train discriminator
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_imgs = X_train[idx]

        noise = np.random.normal(0, 1, (batch_size, 100))
        fake_imgs = generator.predict(noise, verbose=0)

        d_loss_real = discriminator.train_on_batch(real_imgs, real_labels)
        d_loss_fake = discriminator.train_on_batch(fake_imgs, fake_labels)
        
        # Train generator
        noise = np.random.normal(0, 1, (batch_size, 100))
        g_loss = gan.train_on_batch(noise, real_labels)

        if epoch % 2 == 0:
            print(f"Epoch {epoch} | D Loss: {0.5 * np.add(d_loss_real, d_loss_fake)[0]:.4f} | G Loss: {g_loss:.4f}")
            show_generated_images(generator)

# Step 6: Display generated images
def show_generated_images(generator, examples=16):
    noise = np.random.normal(0, 1, (examples, 100))
    generated = generator.predict(noise, verbose=0)
    plt.figure(figsize=(4, 4))
    for i in range(examples):
        plt.subplot(4, 4, i + 1)
        plt.imshow(generated[i], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    print()

# Execute training
train_gan(epochs=10, batch_size=40)
```

## üß† **Experiment 7: Transfer Learning for Image Classification with Pre-trained Models**  
### Transfer learning techniques are applied using pre-trained architectures such as VGG16 or ResNet. The models are fine-tuned on a new image dataset to improve classification performance with fewer training resources.


```python
# Experiment 7: Transfer Learning for Image Classification with Pre-trained Models
# Author: Madhurima Rawat

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing import image_dataset_from_directory
import os, zipfile
import warnings
warnings.filterwarnings('ignore')

# Plot style
sns.set(style="whitegrid", palette="pastel")

print("üß† Experiment 7: Transfer Learning for Image Classification")
print("------------------------------------------------------------")
print("We use the VGG16 model pre-trained on ImageNet.\n")

# Download and extract dataset manually
url = 'https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip'
path_to_zip = tf.keras.utils.get_file('cats_and_dogs_filtered.zip', origin=url, extract=False)

# Unzip it manually
extract_dir = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')
if not os.path.exists(extract_dir):
    with zipfile.ZipFile(path_to_zip, 'r') as zip_ref:
        zip_ref.extractall(os.path.dirname(path_to_zip))

# Set train and validation directories
base_dir = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')

# Load datasets
img_size = (150, 150)
batch_size = 32

train_ds = image_dataset_from_directory(
    train_dir, shuffle=True, image_size=img_size, batch_size=batch_size)

val_ds = image_dataset_from_directory(
    val_dir, shuffle=True, image_size=img_size, batch_size=batch_size)

# Prefetch for performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

# VGG16 base
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
for layer in base_model.layers:
    layer.trainable = False

# Custom head
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(2, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(train_ds, validation_data=val_ds, epochs=3)

# Plot training
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    plt.figure(figsize=(12, 4))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(acc, label='Train Acc', marker='o')
    plt.plot(val_acc, label='Val Acc', marker='o')
    plt.title("üìà Accuracy Over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(loss, label='Train Loss', marker='o')
    plt.plot(val_loss, label='Val Loss', marker='o')
    plt.title("üìâ Loss Over Epochs")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_training_history(history)
```

## üåÄ **Experiment 8: Autoencoders for Data Compression and Denoising**  
### Autoencoders are implemented for tasks like image denoising and data compression. The experiment explores how autoencoders learn compact, informative representations of data and reconstruct inputs effectively.

```python
# Experiment 8: Autoencoders for Data Compression and Denoising
# Author: Madhurima Rawat

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import warnings
warnings.filterwarnings('ignore')

# Plot style
sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üåÄ Experiment 8: Autoencoders for Compression and Denoising")
print("------------------------------------------------------------")
print("This experiment uses a simple autoencoder on the MNIST dataset.")
print("It compresses and reconstructs images, demonstrating dimensionality reduction and denoising.\n")

# Step 1: Load and preprocess data
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), -1))
x_test = x_test.reshape((len(x_test), -1))

# Step 2: Add noise
noise_factor = 0.5
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

# Step 3: Build autoencoder
input_img = Input(shape=(784,))
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(784, activation='sigmoid')(decoded)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Step 4: Train model
autoencoder.fit(x_train_noisy, x_train,
                epochs=10,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test_noisy, x_test),
                verbose=1)

# Step 5: Reconstruct images
decoded_imgs = autoencoder.predict(x_test_noisy)

# Step 6: Visualize original vs. noisy vs. reconstructed
def plot_autoencoder_results(x_noisy, x_decoded):
    n = 10
    plt.figure(figsize=(18, 6))
    for i in range(n):
        # Original noisy
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(x_noisy[i].reshape(28, 28), cmap='gray')
        plt.title("Noisy")
        plt.axis('off')

        # Reconstructed
        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(x_decoded[i].reshape(28, 28), cmap='gray')
        plt.title("Denoised")
        plt.axis('off')

        # Clean Ground Truth
        ax = plt.subplot(3, n, i + 1 + 2*n)
        plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
        plt.title("Original")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
    print()

plot_autoencoder_results(x_test_noisy, decoded_imgs)
```

## üéõÔ∏è **Experiment 9: Hyperparameter Tuning for Deep Learning Models**  
### Various hyperparameters such as learning rate, batch size, and optimizer choice are tuned to observe their effects on model performance. Grid search and random search techniques are utilized for optimization.


```python
# Experiment 9: Hyperparameter Tuning for Deep Learning Models
# Author: Madhurima Rawat

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import ParameterGrid
import warnings
warnings.filterwarnings('ignore')

sns.set(style="whitegrid", palette="pastel")

# -------------------- Introduction --------------------
print("üéõÔ∏è Experiment 9: Hyperparameter Tuning")
print("----------------------------------------")
print("We explore tuning learning rate, batch size, and optimizer.")
print("Using grid search to evaluate combinations on MNIST.\n")

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype("float32") / 255.
x_test = x_test.astype("float32") / 255.
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Define model builder
def build_model(optimizer='adam', lr=0.001):
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer=optimizer, 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])
    return model

# Grid search parameters
param_grid = {
    'batch_size': [64, 128],
    'optimizer': ['adam', 'sgd'],
    'epochs': [5]
}
grid = list(ParameterGrid(param_grid))

# Store results
results = []

# Run grid search
for i, params in enumerate(grid):
    print(f"üîÑ Running config {i+1}: {params}")
    model = build_model(optimizer=params['optimizer'])
    history = model.fit(x_train, y_train,
                        epochs=params['epochs'],
                        batch_size=params['batch_size'],
                        verbose=0,
                        validation_split=0.1)
    val_acc = history.history['val_accuracy'][-1]
    results.append((params, val_acc))
    print(f"‚úÖ Validation Accuracy: {val_acc:.4f}\n")

# Plot results
labels = [str(r[0]) for r in results]
scores = [r[1] for r in results]

plt.figure(figsize=(10, 5))
sns.barplot(x=scores, y=labels, palette="pastel")
plt.title("üìä Validation Accuracy for Hyperparameter Configurations")
plt.xlabel("Accuracy")
plt.ylabel("Configuration")
plt.tight_layout()
plt.show()

print()
```
